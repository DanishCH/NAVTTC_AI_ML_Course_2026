## **Mean, Median, and Mode**

For these statistical tasks, we will use NumPy and SciPy. We will analyze an array representing AI Engineer starting salaries (in thousands of PKR) and an array of model inference response times.

```python
import numpy as np
from scipy import stats

print("--- Example 1: Analyzing Tech Salaries (in '000 PKR) ---")
# Salaries at companies like Careem, Afiniti, Systems Ltd, etc.
salaries = [120, 150, 135, 120, 180, 200, 120, 160]

print(f"Mean (Average Salary): Rs. {np.mean(salaries):.1f}k")
print(f"Median (Middle Salary): Rs. {np.median(salaries)}k")
print(f"Mode (Most Common): Rs. {stats.mode(salaries, keepdims=True).mode[0]}k")

print("\n--- Example 2: Analyzing AI Model Response Times (in milliseconds) ---")
response_times = [45, 50, 42, 55, 45, 300, 48, 45] # Note the 300ms outlier!

print(f"Mean Time: {np.mean(response_times):.1f} ms (Pulled up by the 300ms outlier)")
print(f"Median Time: {np.median(response_times)} ms (Better reflection of normal speed)")
print(f"Mode Time: {stats.mode(response_times, keepdims=True).mode[0]} ms")
```
