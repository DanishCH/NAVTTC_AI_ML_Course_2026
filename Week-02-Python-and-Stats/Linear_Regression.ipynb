# **Linear Regression**

Linear regression finds the line of best fit ($y = mx + c$). Here are two models: one predicting salary based on experience, and another predicting how long an AI model takes to process text based on the word count.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

print("--- Example 1: Predicting Salary from Years of Experience ---")
# X = Years of Experience, y = Salary (in '000 PKR)
X_exp = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_salary = np.array([120, 160, 210, 250, 310])

salary_model = LinearRegression().fit(X_exp, y_salary)
pred_3_5_years = salary_model.predict(np.array([[3.5]]))
print(f"Predicted salary for 3.5 years experience: Rs. {pred_3_5_years[0]:.0f}k")


print("\n--- Example 2: Predicting Processing Time based on Word Count ---")
# X = Number of Words in prompt, y = Processing Time (ms)
X_words = np.array([10, 50, 100, 200, 500]).reshape(-1, 1)
y_time = np.array([20, 80, 150, 290, 700])

time_model = LinearRegression().fit(X_words, y_time)
pred_300_words = time_model.predict(np.array([[300]]))
print(f"Predicted processing time for a 300-word prompt: {pred_300_words[0]:.0f} ms")
```
