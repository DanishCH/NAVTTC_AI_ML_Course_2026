# **Standard Deviation**

Standard deviation ($\sigma$) measures how spread out your data is from the mean ($\mu$). A low standard deviation means the data is tightly clustered.

```python
import numpy as np

print("--- Example 1: High vs. Low Variance in API Latency ---")
# Model A has very consistent response times
model_a_latency = [50, 52, 49, 51, 50]
# Model B has erratic response times
model_b_latency = [10, 90, 20, 100, 30]

print(f"Model A Standard Deviation: {np.std(model_a_latency):.2f} ms (Very stable)")
print(f"Model B Standard Deviation: {np.std(model_b_latency):.2f} ms (Highly unstable)")

print("\n--- Example 2: Analyzing Exam Scores in a Dataset ---")
# Checking the spread of scores in a training dataset
test_scores = [85, 88, 90, 92, 86, 89]
print(f"Scores spread out from the average by roughly: {np.std(test_scores):.2f} points")
```
