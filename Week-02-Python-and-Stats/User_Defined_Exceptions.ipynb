# **User-Defined Exceptions**

Sometimes standard Python errors aren't specific enough. Here, we create our own custom errors for specific AI and data engineering rules.

```python
# Define our custom exception classes
class ModelAccuracyTooLowError(Exception):
    pass

class InvalidSalaryError(Exception):
    pass

print("--- Example 1: Enforcing Machine Learning Standards ---")
def deploy_model(accuracy):
    try:
        if accuracy < 0.85:
            raise ModelAccuracyTooLowError(f"Accuracy {accuracy*100}% is below the 85% deployment threshold.")
        print("Model approved for deployment to production.")
    except ModelAccuracyTooLowError as e:
        print(f"Deployment Blocked: {e}")

deploy_model(0.92) # Will pass
deploy_model(0.78) # Will raise our custom error


print("\n--- Example 2: Validating Data Entry ---")
def record_starting_salary(salary_pkr):
    try:
        if salary_pkr < 0:
            raise InvalidSalaryError("A salary cannot be a negative number.")
        print(f"Salary of Rs. {salary_pkr} recorded successfully.")
    except InvalidSalaryError as e:
        print(f"Data Entry Error: {e}")

record_starting_salary(150000)
record_starting_salary(-50000)
```
