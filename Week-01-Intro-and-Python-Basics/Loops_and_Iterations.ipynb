# **Loops and Iterations**

**For Loops**

```python
# Iterating over a list using a for loop
tools = ["Python", "TensorFlow", "Azure"]
for tool in tools:
    print(f"Learning {tool}")
```

**While Loops**

```python
# Using a while loop to count
count = 0
while count < 3:
    print(f"Count is {count}")
    count += 1
```

**Break Statement**

```python
# Stopping a loop early using break
for i in range(5):
    if i == 3:
        break
    print(i)
```

**Continue Statement**

```python
# Skipping an iteration using continue
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Looping Techniques**

```python
# Using enumerate() to get index and value
algorithms = ["Linear Regression", "Decision Tree", "SVM"]
for index, algo in enumerate(algorithms):
    print(f"Algorithm {index + 1}: {algo}")
```
