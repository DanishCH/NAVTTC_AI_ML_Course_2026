**Taking numerical input for a machine learning parameter**

```python
# Taking numerical input for a machine learning parameter
try:
    epochs = int(input("Enter the number of training epochs (e.g., 50): "))
    print(f"Training will run for {epochs} epochs.")
except ValueError:
    print("Error: Invalid input! Please enter a whole number.")
```
