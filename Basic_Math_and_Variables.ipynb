# **Basic Math and Variables**

**Add two numbers**

```python
# This program adds two numbers
num1 = 1.5
num2 = 6.3

# Add two numbers
sum_result = num1 + num2

# Display the sum
print('The sum of {0} and {1} is {2}'.format(num1, num2, sum_result))
```

**Add two numbers (User Input)**

```python
# Store input numbers
num1 = input('Enter first number: ')
num2 = input('Enter second number: ')

# Add two numbers
sum_result = float(num1) + float(num2)

# Display the sum
print('The sum of {0} and {1} is {2}'.format(num1, num2, sum_result))
```

**Calculate Square Root (Real Number)**

```python
# Python Program to calculate the square root
num = 8

# To take the input from the user
num = float(input('Enter a number: '))

num_sqrt = num ** 0.5
print('The square root of %0.3f is %0.3f' % (num, num_sqrt))
```

**Square Root of Real or Complex Numbers**

```python
# Find square root of real or complex numbers
import cmath # Importing the complex math module

num = 1+2j

# To take input from the user
# num = eval(input('Enter a number: '))

num_sqrt = cmath.sqrt(num)
print('The square root of {0} is {1:0.3f}+{2:0.3f}j'.format(num, num_sqrt.real, num_sqrt.imag))
```

**Celsius to Fahrenheit Conversion**

```python
# Python Program to convert temperature in celsius to fahrenheit
celsius = 37.5

# calculate fahrenheit
fahrenheit = (celsius * 1.8) + 32
print('%0.1f degree Celsius is equal to %0.1f degree Fahrenheit' % (celsius, fahrenheit))
```

**Area of a Triangle**

```python
# Python Program to find the area of triangle
a = 5
b = 6
c = 7

# Uncomment below to take inputs from the user
# a = float(input('Enter first side: '))
# b = float(input('Enter second side: '))
# c = float(input('Enter third side: '))

# calculate the semi-perimeter
s = (a + b + c) / 2

# calculate the area
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
print('The area of the triangle is %0.2f' % area)
```

**Solve Quadratic Equation $ax^2+bx+c=0$**

```python
# Solve the quadratic equation
import cmath

a = 1
b = 5
c = 6

# calculate the discriminant
d = (b**2) - (4 * a * c)

# find two solutions
sol1 = (-b - cmath.sqrt(d)) / (2 * a)
sol2 = (-b + cmath.sqrt(d)) / (2 * a)

print('The solutions are {0} and {1}'.format(sol1, sol2))
```

**Kilometers to Miles Conversion**

```python
# Taking kilometers input from the user (Hardcoded for testing)
# kilometers = float(input("Enter value in kilometers: "))
kilometers = 10.0

# conversion factor
conv_fac = 0.621371

# calculate miles
miles = kilometers * conv_fac
print(f'{kilometers} kilometers is equal to {miles:.2f} miles')
```
